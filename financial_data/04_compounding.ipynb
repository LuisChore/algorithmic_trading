# Discrete Compounding

##  01. Annual Compounding

Interests accrue once a year at the end of the year

* **stated rate**: Specified rate, it does not take into account the effect of 'compounding'
* **effective rate**: True value you earn or pay

$final\;value = value \cdot (1 + stated\_rate)^{years}$




$effective\_annual\_rate = \frac{Final\; Price}{Initial\;Price}^{\frac{1}{years}}-1$

## 02. Quarterly Compounding

Interests accrue once a quarter at the end of the quarter.
 
Quarterly compounding is favourable as we earn compound interest (interest on interest)

$final\;value = value \cdot ( 1 + \frac{stated\_rate}{quarters})^{years \cdot quarters}$


$effective\_annual\_rate = \frac{Final\; Price}{Initial\;Price}^{\frac{1}{years}}-1$

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set()

def final_value(initial_value,rate,n_events,years = 1):
  return initial_value * (1 + (rate/n_events))**(years*n_events)

events = [1,2,3,4,6,12] #(annualy,semesterly,quarterly,trimesterly,bimesterly,monthly)
final_values = [final_value(100,0.10,x) for x in events]

plt.figure(figsize=(10, 6))
plt.plot(events,final_values)
plt.xlabel("Event")
plt.ylabel("Final Price")
plt.show()
```

# Continous Compounding

We can increase the number of compounding events (quarters) to infinite, then we end up with continous compounding



$final\;value = value \cdot ( 1 + \frac{stated\_rate}{quarters})^{years \cdot quarters}$


$ \lim_{quarters\to\infty} : $

$ final\;value = value \cdot e^{years \cdot stated\_rate}$



---



```python
x = np.linspace(1,200)
rate = .10
years = 1
initial_value = 100

y = np.array([final_value(initial_value,rate,i,years) for i in x])
euler = initial_value * np.exp(years * rate)

plt.figure(figsize=(15, 8))
plt.plot(x,y,label = "Price")
plt.xlabel("Event")
plt.ylabel("Final Price")

plt.hlines(euler,color = 'r',xmin = 1,xmax = 200, label = 'inf events')
plt.legend()
plt.show()
```

## 01. Effective Annual Rate

$effective\_annual\_rate = \frac{Final\; Price}{Initial\;Price}^{\frac{1}{years}}-1$

When the frequency of compounding is increased up to infinity the calculation will be: 

$effective\_annual\_rate = e^{rate} - 1$


## 02. Log Returns /Stated Rate

It calculates the stated rate. Log returns assumes returns are compounded continuously rather than across sub-periods

$r = ln(\frac{Final\;Value}{Initial\;Value})$

$ r = ln(effective\_annual\_rate + 1) $
