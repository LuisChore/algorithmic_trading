## 01. Covariance

$ Cov\;Matrix = \begin{bmatrix}
\sigma_1 ^{2} & \sigma_{1,2} ^{2} & \cdots  &\sigma_{1,I} ^{2}  \\
\sigma_{2,1} ^{2}&\sigma_2 ^{2}  & \cdots & \sigma_{2,I} ^{2} \\
\vdots  & \vdots & \ddots  &  \\
 \sigma_{I,1} ^{2}&\sigma_{I,2} ^{2}  &  & \sigma_{I} ^{2} \\
\end{bmatrix} $

* Representation of the way two and more variables relate each other.
* The main diagonal are the variance of each one of the variables.

```python
import numpy as np
import pandas as pd
import pandas_datareader.data as wb
import matplotlib.pyplot as plt 
```

```python
'''
returns the historical price data of a company, including the percentage daily 
change, daily log return
'''
def log_return(ticket,data_source,start_year):
  info = wb.DataReader(ticket,data_source,start = start_year)
  info['log_return'] = np.log((info['Adj Close'] / info['Adj Close'].shift(1)))
  return info
```

### Annual Covariance Matrix

```python
'''
Calculates the ANNUAL COVARIANCE MATRIX of different tickets using the 
logarithmic rate of returns
@TODO: check that all the companies have the same amount of returns              
'''
def annual_cov_matrix(tickets,start_year):
  data = pd.DataFrame()
  for ticket in tickets:
    data[ticket] = log_return(ticket,'yahoo',start_year)['log_return']
  cov_matrix = data.cov() * 250 #annualizing 
  return cov_matrix
```

```python
tickets = ['PG','MSFT','F','GE']
start_year = 2005
annual_cov_matrix(tickets,start_year)
```

## 02. Correlation

Correlation between **returns of the companies**.

It reflects the dependence between prices at different times and focuses on the returns of your portfolio. It is not necessary annualize the result, it shows only the relation between the two variables 

```python
'''
@TODO: check that all the companies have the same amount of returns      
'''
def corr_matrix(tickets,start_year):
  data = pd.DataFrame()
  for ticket in tickets:
    data[ticket] = log_return(ticket,'yahoo',start_year)['log_return']
  corr_matrix = data.corr() #not necessary annualizing
  return corr_matrix
```

```python
tickets = ['PG','MSFT','F','GE']
start_year = 2005
corr_matrix(tickets,start_year)
```

### Example with different number of returns

The correlation function ignore null values, it is ok if one of the companies does not have old data

```python
tickets = ['MSFT','UBER']
corr_matrix(tickets,'2005')
```

```python
tickets = ['MSFT','UBER']
corr_matrix(tickets,'2019-05-10')
```
